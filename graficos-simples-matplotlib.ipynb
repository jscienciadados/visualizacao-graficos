# Grafico de Barras

```
# definindo os dados
notas = [6, 4, 8, 9, 6, 10]
pesos = [3, 5, 2, 4, 6, 3]

plt.bar(notas, pesos, label = 'Gr Barras', color = 'r')
plt.xlabel('Notas')
plt.ylabel('Pesos')
plt.legend()
plt.show()
```

```
x1 = [2, 4, 6, 8, 10]
y1 = [6, 7, 8, 2, 4]
x3 = [1, 3, 5, 7, 9]
y3 = [6, 7, 8, 2, 4]

plt.bar(x1, y1, label = 'Barras 1', color = 'r')
plt.bar(x3, y3, label = 'Barras 2', color = 'y')
plt.legend()
plt.show()
```

# Stack Plots

```
dias = [1, 2, 3, 4, 5]
dormir = [7, 8, 6, 77, 7]
comer = [2, 3, 4, 5, 3]
work = [7, 8, 7, 2, 2]
lazer = [8, 5, 7, 8, 13]
```

```
plt.stackplot(dias, dormir, comer, work, lazer, colors= ['m', 'c', 'r', 'k', 'b'])
plt.show()
```

# Pie Chart

```
fatias = [7, 2, 2, 13]
atividades = ['dormir', 'comer', 'worker', 'lazer']
colunas = ['c', 'm', 'r', 'k']
```

```
plt.pie(fatias, labels= atividades, colors= colunas, startangle= 90, shadow= True, explode= (0,0.1,0,0))
plt.show()
```

# Gráficos

```
# O pylab combina funcionalidades do pyplot com as funcionalidade do numpy
from pylab import *
```

```
x = linspace(0, 5, 10)
y = x ** 2
# definindo uma figura
fig = plt.figure()
# definindo os eixos
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Grafico de Linha')
```

```
# Grafico com 2 figuras
x = linspace(0, 5, 10)
y = x ** 2
# definindo uma figura
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

# figura principal
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Figura Principal')

# figura Secundaria
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('Figura Secundaria');


```

```
# Grafico em Parelo 
fig, axes = plt.subplots(nrows= 1, ncols= 2)
for ax in axes:
   ax.plot(x,y, 'r')
   ax.set_label('x')
   ax.set_ylabel('y')
   ax.set_title('Título')
```

```
plt.scatter(np.arange(50), np.random.randn(50))
plt.show()
```

```
# Plot e Scatter 
fig = plt.figure(figsize=(15,5))

ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(np.random.randn(50), color='red')

ax2 = fig.add_subplot(1, 2, 2)
ax2.scatter(np.random.randn(50), np.random.randn(50))
plt.show()
```

```
# Plots diversos
_, ax = plt.subplots(2,3)

ax[0,1].plot(np.random.randn(50), color = 'green', linestyle = '-')
ax[1,0].hist(np.random.randn(50))
ax[1,2].scatter(np.random.randn(50), np.random.randn(50), color = 'red')
plt.show()
```

```
# Diferentes tipos de plots
xx = np.linspace(-0.75, 1., 100)
n = np.array([0, 1, 2, 3, 4, 5])

fig, axes = plt.subplots(1, 4, figsize=(12,3))

axes[0].scatter(xx, xx + 0.25 * randn(len(xx)))
axes[0].set_title('Scatter')

axes[1].step(n, n**2, lw=2)
axes[1].set_title('Step')

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[2].set_title('Bar')

axes[3].fill_between(xx, xx**2, xx**3, color="green", alpha=0.5);
axes[3].set_title('fiil_between');
```

```
# Color Map
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def ColorMap(phi_m, phi_p):
    return ( + alpha - 2 * np.cos(phi_p)*cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p))

phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X,Y = np.meshgrid(phi_p, phi_m)
Z = ColorMap(X, Y).T

fig, ax = plt.subplots()

p = ax.pcolor(X/(2*np.pi), Y/(2*np.pi), Z, cmap=cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max())
cb = fig.colorbar(p, ax=ax)    
```
